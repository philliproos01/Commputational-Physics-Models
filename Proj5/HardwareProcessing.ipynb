```python
#This is a simple program made to model the COLLECTED data from our hardware exported to a csv file

import matplotlib.pyplot as plt
import math
import csv
  
x = []
humidity = []
y = []
pressure = []
dpt = []
  
with open('Arduino.csv','r') as csvfile:
    lines = csv.reader(csvfile, delimiter=',')
    for row in lines:
        b = 17.62; #these are constants
        c = 243.5;

        gamma = math.log(((float(row[2]))/100) + b*(float(row[1])) / (c+(float(row[1])))); 
        tdp = c*gamma / (b-gamma);
        #tdpf = (tdp*1.8) + 32;
        x.append(float(row[0]))
        y.append(float(row[1]))
        humidity.append(float(row[2]))
        pressure.append(float(row[3]))
        dpt.append(tdp)
        
#plt_1 = plt.figure(figsize=(16, 14))
fig, ax1 = plt.subplots(figsize=(16, 16))
ax2 = ax1.twinx()  
#ax2.set_ylim(4, 20);


ax2.plot(x, y, color = 'blue', label = "Temperature", lw=4)
ax1.plot(x, humidity, color = 'red', label = "Humidity", lw=4)
ax2.plot(x, dpt, color = 'green', label = "Dew point", lw=4)


#ax1.plot(x, dpt, color = 'green', label = "Dew Point", lw=4)
#ax1.xticks(rotation = 90)
ax1.set_xlabel('Time past in minutes')
ax2.set_ylabel('Temperature(C)')
ax1.set_ylabel('Humidity')
plt.title('Weather Report', fontsize = 20)
plt.grid()
plt.legend()


#plt.figure().set_figwidth(1200)
#plt.figure().set_figwidth(1200)
plt.show()
```
