```python
#This is a simple program made to model the COLLECTED data from the coolant test rig to a csv file

import matplotlib.pyplot as plt
import math
import csv
  
time = []
temperatureCH2 = []
temperatureCH3 = []

pressureCH4 = []
pressureCH5 = []
pressureCH6 = []
pressureCH7 = []
GPMCH1 = []
humidity = []

with open('230811-155719.csv','r') as csvfile:    #put the file name you wish to test here
    lines = csv.reader(csvfile, delimiter=',')
    
    for row in range(33):
        next(csvfile)
    for row in lines:
        time.append(float(row[0]))
        GPMCH1.append(float(row[3][3:]))
        
        temperatureCH2.append(float(row[4][3:]))
        temperatureCH3.append(float(row[5][3:]))
        pressureCH4.append(float(row[6][3:]))
        pressureCH5.append(float(row[7][3:])) 
        pressureCH6.append(float(row[8][3:]))
        pressureCH7.append(float(row[9][3:]))

fig, ax1 = plt.subplots(figsize=(16, 16))
ax2 = ax1.twinx()  

ax1.set_ylim([0, 70])
ax2.set_ylim([0, 10])
ax1.plot(time, temperatureCH2, color = 'red', label = "Temperature CH2", lw=4)
ax1.plot(time, temperatureCH3, color = 'orange', label = "Temperature CH3", lw=4)
ax2.plot(time, pressureCH4, color = 'blue', label = "Pressure CH4", lw=4)
ax2.plot(time, pressureCH5, color = 'green', label = "Pressure CH5", lw=4)
ax2.plot(time, pressureCH6, color = 'pink', label = "Pressure CH6", lw=4)
ax2.plot(time, pressureCH7, color = 'black', label = "Pressure CH7", lw=4)


ax1.set_xlabel('Time (s)')
ax1.set_ylabel('Temperature (C)')
ax2.set_ylabel('Pressure')
plt.title('Temperature and Pressure Changes', fontsize = 20)
plt.grid()
plt.legend()
plt.show()

fig, ax = plt.subplots(figsize=(16, 16))
ax.plot(time, GPMCH1, color = 'blue', label = "GPM", lw=4)
plt.title('Gallons Per Minute (GPM)', fontsize = 20)
plt.grid()
plt.legend()
plt.show()
```
